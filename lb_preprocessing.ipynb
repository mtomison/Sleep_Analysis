```python
import pandas as pd
from datetime import timedelta
```

```python
file_path = "Resources/sleepdata.csv"
sleepData = pd.read_csv(file_path)
sleepData
```

```python
sleepData.isnull().sum()
```

```python
sleepData["Wake up"].count()
```

```python
sleepData["Heart rate"].count()
```

```python
#Drop Wake up and Heart rate columns because more than 70% of each is NaN
sleepData.drop("Wake up", axis = 1, inplace = True)
sleepData.drop("Heart rate", axis = 1, inplace = True)
sleepData
```

```python
sleepData.count()
```

```python
sleepData.dtypes
```

```python
#Replace : with . in order to parse column to float
sleepData["Time in bed"] = sleepData["Time in bed"].str.replace(':', '.')
```

```python
sleepData["Time in bed"] = pd.to_numeric(sleepData["Time in bed"])
```

```python
#Convert Time in bed column from hrs.mins to mins
sleepData["Time in bed"] = sleepData["Time in bed"] * 60
```

```python
#Replace % with '' in order to parse column to int
sleepData["Sleep quality"] = sleepData["Sleep quality"].str.replace('%', '')
```

```python
sleepData["Sleep quality"] = pd.to_numeric(sleepData["Sleep quality"])
```

```python
#Divide percentages by 100 to get all vaules between 0 and 1
sleepData["Sleep quality"] = sleepData["Sleep quality"] / 100
```

```python
#Convert Start and End times to datetime
sleepData['Start'] = pd.to_datetime(sleepData["Start"])
sleepData['End'] = pd.to_datetime(sleepData["End"])
```
